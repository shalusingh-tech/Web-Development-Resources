Ans) An internet is an  **`network of electronic devices like computer, mobile,etc. connected to one another forming web like structure`** only for the purpose of ***sharing resources*** all over the world.

Ans) A web browser is a software or tool, that helps to **visualize the website**.

It is a ***part of a internet allow users to `enter web address or search terms to access website`***

**Example:-** Some popular web browser are **`Google Chrome, Mozilla Firefox, Safari, etc.`**

Ans) A browser engine is also a software it use for **`interpreting web content and rendering it visually on your screen.`**

Modern day web engine is divided into two parts:-

1)**`Layout or rendering Engine`**:= It handle **HTML and CSS file** and responsible for showing content along with design of a website.

2)**`Javascript Engine`**:= It handle **Javasript file** which provide dynamic ***interaction between user and website***  

Ans) A Website whose ***overall data is not generated dynamically from server*** and `the data doesn't change frequently over time.`

**In static website the content of the site is not going to change a lot.**

***A static site is built using only `HTML,CSS and Javascript`, and `doesn't require any server site processing.`***

They are **`provide simplicity,are faster, cost effective and scalable`**

**Example:-**`Consider you are writing a blog related to your newly developed project.So in case like this where your content is not going to change a lot over time we simply use static website.`

Ans) In dynamic site **`data changes frequently`**. It uses `server-side programming and database to generate its content`.

It provide **`dynamic and interactive function to user.`**

It has ability like ***`searching data in database, submit form data, display customer content ,etc.`***

Dynamic site provide **flexibility,Better SEO , advanced functionality, etc.**

**Examplt:-** `Facebook, instagram, youtube`, etc.
